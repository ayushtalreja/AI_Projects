```python
car_sales_missing = pd.read_csv("9.2 car-sales-missing-data.csv")
car_sales_missing
```

```python
# important to reassign
car_sales_missing["Make"] = car_sales_missing["Make"].str.lower()
car_sales_missing
```

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

```python
car_sales_missing["Odometer"]=car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean()) 
car_sales_missing
```

```python
car_sales_missing["Doors"].fillna(car_sales_missing["Doors"].mean(),inplace=True)
car_sales_missing
```

```python
#Assigning one value to all
car_sales_missing["Seats"] = 4
car_sales_missing
```

```python
#Adding a column through Series dataType
Fuel_Consumed = pd.Series([500,300,120,125,12121,11145,4545,11,65,48])
car_sales_missing["Fuel_Consumed(L)"] = Fuel_Consumed
car_sales_missing
```

```python
#Adding a column through Python list
mileage = [18,20,30,51,12,14,23,17,25,56]
car_sales_missing["Mileage"] = mileage
car_sales_missing
```

```python
car_sales_missing.drop("Fuel_Consumed(L)",axis= 1,inplace=True)
```
