## Introduction to matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# shows a figure but also shows an empty braces
#There are two ways of plotting in matplotlib pyplot and oo version.This one is pyplot version.
plt.plot()
```

```python
#semi colon removes the braces
plt.plot();
```

```python
#Another way of showing a figure without that empty braces
plt.plot()
plt.show()
```

```python
plt.plot([1,2,3,4],[2,4,6,8]);
```

```python
x = [2,3,4] # x and y are being used later
y = [11,22,33]
plt.plot(x,y);
```

```python
#Now we will be discussing OO version of plotting.There are three ways in this one
```

```python
fig = plt.figure() # create a figure
ax = fig.add_axes([1,1,1,1])  #add axes, rect = [left, bottom,width, height] 
ax.plot(x,y)  #then plot
plt.show()
```

```python
# 2nd way
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
#3rd way
fig,ax = plt.subplots()
ax.plot([6,9,12],y)
plt.show() # removes those matplotlib caption
```

## Subplots


```python
# option 1 without indexing
fig,((ax0,ax1), (ax2,ax3)) = plt.subplots(ncols=2,
                                         nrows=2
                                         ,figsize=(10,10))
fig.tight_layout(pad=8)
ax0.plot(array_1,np.exp(array_1));
ax1.barh(list(Clothes_Prices.keys()),list(Clothes_Prices.values()));
ax2.hist(hist_data);
ax3.scatter(array_1,np.exp(array_1));
```

```python
# Option 2 with indexing
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,7))
ax[0,0].plot(array_1,np.exp(array_1))
ax[0,1].bar(Clothes_Prices.keys(),Clothes_Prices.values())
ax[1,0].hist(hist_data)
ax[1,1].scatter(array_1,np.exp(array_1))
```

## ploting with pandas using oo method

```python
# up until now we were plotting using the pyplot pandas wrapper which is quick but has less flexibility
heart_disease.head()
```

```python
type(heart_disease["age"][0])
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
```

```python
over_50
```

```python
#Scatter plot of age against cholestrol
over_50.plot(
    x="age",
    y="chol",
    kind="scatter",
    c="target",
    figsize=(10,6));
# Here you cant define x axis label
```

```python
## Mixed pyplot and OO method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(x="age",
                   y="chol",
                   kind="scatter",
                   c="target",
                   ax=ax)
# Setting x limit
ax.set_xlim([45,100])
```

```python
# Using OO method
fig,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(over_50["age"],over_50["chol"],c=over_50["target"]) 
ax.legend(*scatter.legend_elements(),title="target") # * only removes matplotlib subline 
ax.set(xlabel= "Age",ylabel="Cholestrol levels",title="Heart-Disease Over Age 50");
```

```python
fig,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(over_50["age"],over_50["chol"],c=over_50["target"]) 
ax.legend(*scatter.legend_elements(),title="target")
ax.set(xlabel= "Age",ylabel="Cholestrol levels",title="Heart-Disease Over Age 50")
# Adding horizontal line to the y axis 
ax.axhline(over_50["chol"].mean(),ls="--",c="r");
# ls = linestyle ,c= color
```

## Adding another plot to existing one

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,figsize=(15,10),sharex=True) 
# Sharex will share x label that means one label for the subplots.You can choose where to display it though

scatter0=ax0.scatter(over_50["age"],over_50["chol"],c=over_50["target"])
scatter1=ax1.scatter(over_50["age"],over_50["thalach"],c=over_50["target"])

ax0.legend(*scatter0.legend_elements(),title="Target")
ax1.legend(*scatter1.legend_elements(),title="Target",loc="lower right")

ax0.set(ylabel="Cholestrol",title="Cholestrol over the age 50 in heart disease patients")
ax1.set(xlabel= "Age",ylabel="Max heart rate",title="Heart rate levels over the age 50 in heart disease patients");

ax0.axhline(over_50["chol"].mean(),ls="--",c="b",label="Average")
ax1.axhline(over_50["thalach"].mean(),linestyle="--",c="r",label="Average")

# Adding super title
fig.suptitle("Heart Rate Disease Analysis",fontsize=20,fontweight="bold");
```
